```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5,5,100)
y = (-2*x-3)/4
plt.plot(x, y, '-r', label='2x+4y+3=0')
plt.plot(1,0,'ro')
plt.text(1,0,'A(1,0)')
plt.plot(0,-2,'bo')
plt.text(0,-2,'B(0,-2)')
plt.xlabel('x', color='#1C2833')
plt.ylabel('y', color='#1C2833')
plt.grid()
plt.show()
```
